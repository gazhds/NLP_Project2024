# Looking into the mislabeled spans

```python
def read_iob2_file(path):
    
    data = []
    current_words = []
    current_tags = []

    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()#[2:] ##changed this
        
    for line in lines:
        
        line = line.strip()
        
        if line: # if line is not an empty line
            tok = line.split('\t')
            current_words.append(tok[1])
            current_tags.append(tok[2])
            
        else:
            if current_words:
                data.append((current_words, current_tags))
            current_words = []
            current_tags = []
            
            
    if current_tags != []:
        data.append((current_words, current_tags))

    df = pd.DataFrame(data, columns=['words', 'tags'])
    df['id'] = df.index
    df = df[['id', 'words', 'tags']]
    
    return df
```

```python
def read_bio_file(path):
    
    data = []
    current_words = []
    current_tags = []

    with open(path, "r", encoding="utf-8") as f:
        lines = f.readlines()[2:] 
        
    for line in lines:
        
        line = line.strip()
        
        if line: # if line is not an empty line
            tok = line.split('\t')
            current_words.append(tok[0])
            current_tags.append(tok[3])
            
        else:
            if current_words:
                data.append((current_words, current_tags))
            current_words = []
            current_tags = []
            
            
    if current_tags != []:
        data.append((current_words, current_tags))

    df = pd.DataFrame(data, columns=['words', 'tags'])
    df['id'] = df.index
    df = df[['id', 'words', 'tags']]
    
    return df
```

```python
PATH_TRUE = "datasets/10p_test.bio"
PATH_PRED = "preds_10p/pred_10p_on_10p.iob2"
true_labels = read_bio_file(PATH_TRUE)
predicted_labels = read_iob2_file(PATH_PRED)
```

```python
all_true = []
all_pred = []
for prediction in true_labels["tags"]:
    for label in prediction:
        all_true.append(label)
for prediction in predicted_labels["tags"]:
    for label in prediction:
        all_pred.append(label)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
labels_list = list(set(all_true))
labels_list.sort()
cm = confusion_matrix(y_true=all_true, y_pred=all_pred, normalize='true',labels=labels_list)
```

```python
cm = np.round(cm, decimals=2)
```

```python
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels_list)
fig, ax = plt.subplots(figsize=(10, 10))
disp.plot(ax=ax, cmap=plt.cm.Blues)

# Rotate the tick labels 
plt.xticks(rotation=45, ha="right")

# Adjust layout
plt.tight_layout()

plt.show()
```
